```python
# for K fold
X_k = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
y_k = np.array([[1, 2, 3, 4]])
print(X_k.shape)

t_k = np.concatenate((X_k, y_k.T), axis=1)
print(t_k[:,:2])
kf = KFold(n_splits=3)

print(kf.get_n_splits(X_k))
# print(X_k)
for k_train, k_test in kf.split(X_k):
    
    print(X_k[k_train].shape, type(X_k[k_train]))
```

```python
# Parameters
learning_rate = 0.1
training_epochs = 700
batch_size = 9
display_step = 100


# Network Parameters
n_hidden_1 = 50 # 1st layer number of features
n_hidden_2 = 50 # 2nd layer number of features
n_input = 50 # Number of feature
n_classes = 1 # Number of classes to predict


# tf Graph input and output
x = tf.placeholder("float", [None, n_input])
y = tf.placeholder("float", [None, None])

# Create model
def multilayer_perceptron(x, weights, biases):
    
    print("x.shape", x.shape)
    print("weights", weights)
    
    # Hidden layer with RELU activation : changed to sigmoid
    layer_1 = tf.add(tf.matmul(x, weights['w1']), biases['b1'])
    layer_1 = tf.nn.sigmoid(layer_1)
    
    # Hidden layer with RELU activation :  changed to sigmoid
    layer_2 = tf.add(tf.matmul(layer_1, weights['w2']), biases['b2'])
    layer_2 = tf.nn.sigmoid(layer_2)
    
    # Output layer with linear activation 
    out_layer = tf.matmul(layer_2, weights['out']) + biases['out']
    out_layer = tf.nn.sigmoid(out_layer)
    
    return out_layer
    
# Store layers weight & bias
weights = {
    'w1': tf.Variable(tf.random_normal([n_input, n_hidden_1])),
    'w2': tf.Variable(tf.random_normal([n_hidden_1, n_hidden_2])),
    'out': tf.Variable(tf.random_normal([n_hidden_2, n_classes]))
}

biases = {
    'b1': tf.Variable(tf.random_normal([1, n_hidden_1])),
    'b2': tf.Variable(tf.random_normal([1, n_hidden_2])),
    'out': tf.Variable(tf.random_normal([1, n_classes]))
}

print("w1", weights['w1'])
print("b1", biases['b1'])
print("w2", weights['w2'])
print("b2", biases['b2'])

# Construct model
pred = multilayer_perceptron(x, weights, biases)
pred_b = pred > 0.5

# Define loss and optimizer
cost = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=pred, labels=y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)

# Initializing the variables
init = tf.global_variables_initializer()
```

```python
data_kv = np.append(X, Y, axis=1)
print(data_kv.shape, type(data_kv))
kf = KFold(n_splits=10)
KV_data = kf.split(data_kv)
# print(kf.get_n_splits(X_k))
# print(data_kv[:,50])
for k_train, k_test in kf.split(data_kv):
#     print(k_train, k_test)
#     print("K train:", k_train[:,50:].shape)
    print(data_kv[k_train][:,50:].shape, type(data_kv[k_train]), len(data_kv[k_train]), data_kv[k_train][1,50:], data_kv[1,1])

accuracy_train = dict()
accuracy_test = dict()
k_count = 1
```

```python
# Edited by Hasan
# Launch the graph
batch_size = 9
for k_train, k_test in KV_data:
    with tf.Session() as sess:
        sess.run(init)
        # Training cycle
        for epoch in range(training_epochs):
            avg_cost = 0
            total_batch = int(len(k_train)/batch_size) # batch size is 9, 9 * 93 = 837 
            X_batches = np.array_split(k_train[:,:50], total_batch)        
            Y_batches = np.array_split(k_train[:,50:], total_batch)

            # Loop over all batches
            for i in range(total_batch):
                batch_x, batch_y = X_batches[i], Y_batches[i]
                # Run optimization op (backprop) and cost op (to get loss value)
    #             print("batch_x", batch_x)


    #             print("w1", weights['w1'].shape)
    #             print("b1", biases['b1'].shape)
    #             print("w2", weights['w2'].shape)
    #             print("b2", biases['b2'].shape)
    #             print("out_w", weights['out'].shape)
    #             print("out_b", biases['out'].shape)

    #             print("out", sess.run(weights['out']))
    #             outlayer_pred = pred.eval(feed_dict = {x: batch_x})
    #             print("outlayer_pred", outlayer_pred)
    #             print("max", outlayer_pred.max())
    #             print("min", outlayer_pred.min())


                _, c = sess.run([optimizer, cost], feed_dict={x: batch_x,
                                                              y: batch_y})

    #             print("cost c is ", c)
                # Compute average loss
                avg_cost += c / total_batch
            # Display logs per epoch step
            if epoch % display_step == 0:
                print("Epoch:", '%04d' % (epoch+1), "cost=", "{:.9f}".format(avg_cost))
        print("Optimization Finished!")




        # Training accuracy
        X_train = k_train[:,:50]
        Y_train = k_train[:,50:]


        # Test model on training data
        correct_prediction = tf.equal(tf.cast(pred_b, "float"), y)
    #     print("correct prediction", correct_prediction.eval({x: X_train, y: Y_train}))
        # Calculate accuracy
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float32"))
        print("Train Accuracy:", accuracy.eval({x: X_train, y: Y_train}))
        # global result 
    #     result = tf.argmax(pred, 1).eval({x: X_train, y: Y_train})


        # Test accuracy
        # Test model on test data

        correct_prediction = tf.equal(tf.cast(pred_b, "float32"), y)
    #     print("correct prediction", correct_prediction.eval({x: test_X, y: test_Y}))
        # Calculate accuracy
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float32"))
        print("Test Accuracy:", accuracy.eval({x: k_test[:,:50], y: k_test[:,50:]}))
        # global result 
    #     result = tf.argmax(pred, 1).eval({x: test_X, y: test_Y})
```
